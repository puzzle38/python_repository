## 결측치 처리 2

- 날씨는 바로 직전, 직후 날씨와 유사성을 보일 것이기 때문에 이를 활용하여 채워 넣는다.

```python
# origin과 time_hour을 기준으로 정렬해서
# ffill 앞에거를 기준으로 결측치 채우기
# bfill 뒤에거를 기준으로 결측치 채우기

df_weather = df_weather.sort_values(["origin", "time_hour"]).groupby("origin").apply(lambda x : x.fillna(method = "ffill")) \
                    .groupby("origin").apply(lambda x : x.fillna(method = "bfill"))
```

## 1-3 Data join
join 진행
### 1. df_flights는 df_airports와 join : df_flights["dest"], df_airports["faa"] 혹은 df_flghts["origin"], df_flights["faa"] 가능하나, origin은 공항수가 적기 때문에 dest에 대해서만 join을 진행

flights 정보에 airports(공항) 세부 정보를 결합

### 2. 1번 결과와 df_planes join : 1번 df["tailnum"], df_planes["tailnum"]

flight 정보에 planes(항공기) 세부 정보를 결합

### 3. 2번 결과와 df_weather 2번 df[["origin", "time_hour"]], df_weather[["origin", "time_hour"]]

flgiht 정보에 출발한 지역의 날씨와 도착한 지역의 날씨 정보를 결합

```python
Image(path_lecture + "/img/jointypes.png")
```

```python
# 1. df_fligths는 df_airports와 join
# origin에 대해서는 join 하지 않음 -> 개인의 판단에 따라 진행할 수도 있음
df_flights_and_airport = pd.merge(df_flights, df_airports, how = "left", left_on="dest", right_on = 'faa')
df_flights_and_airport.head(2)
```

```python
# 2. 1번 결과와 df_planes join
# 동일한 column name이 있으면 suffixes를 통해 이름을 다르게 join 가능
df_f_a_p = pd.merge(df_flights_and_airport, df_planes, how = 'left', on = 'tailnum', suffixes=("_flghts", "_planes"))
df_f_a_p.head(2)
```

```python
# 3. 2번 결과와 df_weather join
# 2가지 정보에 대한 join 가능
# 1) 출발 시간의 날씨 정보
# 2) 도착 시간의 날씨 정보 : 주어진 자료로 구하기 모호해 진행하진 않음. 만약 가능하다면 진행하는 것도 좋음

# 3-1) 출발 시간의 날씨 정보
df_f_a_p["dest_datetime"] = pd.to_datetime(df_f_a_p["time_hour"])

# time hour는 출발 예정 시간을 반영하기 때문에, 이 데이터에 실제 delayed departure 시간을 더해줘야 한다.
# 또한, time_hour는 '분' 시간이 제외된 데이터이기 때문에, minutes 컬럼 데이터도 더해줘야 한다.
df_f_a_p["dest_datetime"] = df_f_a_p.apply(lambda x : x.dest_datetime + pd.Timedelta(minutes=x.dep_delay) +
                                           pd.Timedelta(minutes = x.minute), axis = 1)
```

```python
df_f_a_p['time_hour']
```

```python
df_f_a_p['dest_datetime']
```

```python
# weather와 join 할 수 있게 dest_datetime 데이터에서 분 정보를 제거
df_f_a_p["dest_datetime"] = df_f_a_p["dest_datetime"].dt.strftime("%Y-%m-%d %H:00:00")
df_f_a_p.head(1)
```

```python
# join을 위해 df_weather의 time_hour를 dest_datetime으로 변경
df_weather = df_weather.rename(columns = {'time_hour' : 'dest_datetime'})
```

```python
df_final = pd.merge(df_f_a_p, df_weather, how = 'left', on = ['origin', 'dest_datetime'], suffixes = ("", "_weather"))
```

```python
df_final.head()
```

```python
# kruskall 검정
# 비모수 검정이기 때문에 정확도가 높진 않음
list_meaningful_column_by_kruskall = []
for column_name in list_categorical_columns:
    list_by_value = []
    for value in df_final[column_name].dropna().unique():
        df_tmp = df_final[df_final[column_name] == value][target_column].dropna()
        list_by_value.append(np.array(df_tmp))
    statistic, pvalue = kruskal(*list_by_value)
    if pvalue <= 0.05:
        list_meaningful_column_by_kruskall.append(column_name)
    print(column_name, ", ", statistic, ", ", pvalue)
print("all categorical columns : ", len(list_categorical_columns))
print("selected columns by kruskall : ", len(list_meaningful_column_by_kruskall), list_meaningful_column_by_kruskall)

# insight : 모든 categorical 변수가 종속변수와 유의미한 관계를 갖는다.
```
